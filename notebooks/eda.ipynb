```python
# Import standard libraries
import pandas as pd               # For loading and manipulating tabular data
import matplotlib.pyplot as plt   # For plotting charts
import seaborn as sns             # For prettier statistical plots
```

```python
# Load the dataset
df = pd.read_csv("../data/raw/customer_churn.csv")

# Show top 5 rows of the dataset
df.head()
```

```python
# Check column info, data types, and missing values
df.info()
```

```python
# Count missing values in each column
df.isnull().sum()
```

```python
# Basic statistical summary of numerical features
df.describe()
```

```python
# View distribution of target column
df['Churn'].value_counts().plot(kind='bar')
plt.title("Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Count")
plt.show()
```

```python
# Correlation heatmap (for numeric features)
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Between Features")
plt.show()
```

```python
# Count plot of churn rate by InternetService
sns.countplot(data=df, x='InternetService', hue='Churn')
plt.title("Churn by Internet Service Type")
plt.show()
```
